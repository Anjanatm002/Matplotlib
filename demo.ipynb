```python
Velocity = [5, 10, 15, 20, 25, 30, 35, 40]
Displacement = [100, 200, 300, 30, 500, 600, 70, 800]
plt.plot(Velocity,Displacement)
plt.xlabel('Velocity',fontstyle='italic',size=13,color="b")
plt.ylabel('Displacement',fontstyle='italic',size=13,color="b")
plt.title('Velocity vs Displacement',fontstyle="normal",size=30,color="r")
plt.grid()
plt.show()
```

```python
Velocity = [5, 10, 15, 20, 25, 30, 35, 40]
Displacement = [100, 200, 300, 30, 500, 600, 70, 800]
plt.plot(Velocity,Displacement)
plt.xlabel('Velocity',fontstyle='italic',size=13,color="b")
plt.ylabel('Displacement',fontstyle='italic',size=13,color="b")
plt.title('Velocity vs Displacement',fontstyle="normal",size=30,color="r")
plt.grid(axis="x",linestyle="dashed",color="r")
plt.show()
```

```python
Velocity = [5, 10, 15, 20, 25, 30, 35, 40]
Displacement = [100, 200, 300, 30, 500, 600, 70, 800]
plt.plot(Velocity,Displacement,marker="s",ms=5,mfc="y",mec="b",linewidth=1,linestyle="solid")
plt.xlabel('Velocity',fontstyle='italic',size=13,color="b")
plt.ylabel('Displacement',fontstyle='italic',size=13,color="b")
plt.title('Velocity vs Displacement',fontstyle="normal",size=30,color="r")
plt.grid(axis="y",linestyle="dotted",color="r")
plt.show()
```

```python
#subplot()
x1=[1,2,3,4]
y1=[3,7,1,9]
plt.subplot(1,2,1)#3rd==graph position,#1st==row,#2nd==column
plt.plot(x1,y1,marker="o")
plt.xlabel('x axis',fontstyle='italic',size=13,color="b")
plt.ylabel('y axis',fontstyle='italic',size=13,color="b")
plt.title('graph1',fontstyle="normal",size=15,color="r")
plt.grid(axis="y",linestyle="dotted",color="r")


x2=[3,4,6,2]
y2=[1,7,2,8]
plt.subplot(1,2,2)
plt.plot(x2,y2,marker="o")
plt.xlabel('x axis',fontstyle='italic',size=13,color="b")
plt.ylabel('y axis',fontstyle='italic',size=13,color="b")
plt.title('graph2',fontstyle="normal",size=15,color="r")
plt.grid(axis="y",linestyle="dotted",color="r")
plt.suptitle("Graph1-Graph2",color="b",size=18)#supertitle


```

```python
#bar chart
x=[1,2,3,4,5,6,7,8]
y=[4,3,6,5,8,2,9,7]
plt.bar(x,y,color="r")
```

```python
country=['uk','india','aus' ,'pak' ,'nz' , 'eng' , 'wi']
trop_count=[1,7,9,5,3,2,6]
plt.bar(country, trop_count,color="y")
plt.xlabel("country name")
plt.ylabel("Tropy_count")
plt.title("Graph_1")
```

```python
country=['uk','india','aus' ,'pak' ,'nz' , 'eng' , 'wi']
trop_count=[1,7,9,5,3,2,6]
plt.barh(country, trop_count,color="y")
plt.xlabel("country name")
plt.ylabel("Tropy_count")
plt.title("Graph_1")
```

```python
#pie chart
#works based on percentage
data1=["Orange","Mango", "Grape","Apple"]
data=[10,20,30,40]
plt.pie(data,labels=data1,autopct="%.1f%%")
plt.legend(loc="upper left")
plt.tight_layout()# to reduce overlapping
```
