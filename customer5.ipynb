```python
#customer5.txt
#1. country count with visualization
import numpy as np
import pandas as pd
```

```python
headers = ["CustomerID", "FName","LName", "Age", "Profession", "Country","Salary"]
df = pd.read_csv("customer5.txt", names=headers, header=None)
df
```

```python
data=df["Country"].value_counts()
data
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,8))
plt.pie(data,labels=data.index,autopct="%.1f%%")
plt.legend(loc="upper right")
plt.tight_layout()
plt.title("Distribution of Country")
```

```python
plt.bar(data.index,data.values)

```
